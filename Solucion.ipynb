### Problema 1:  
Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la función debe devolver una imagen la cual tenga activos los canales de color según los siguientes puntos:
- Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color azul.
- Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde.
- Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color rojo.
- Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y verde.
- Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y azul.
- Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y rojo.

```python
# Importamos librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos función
def activarCanales(varImg, color):

    # Obteniendo dimensiones
    alto = varImg.shape[0]
    ancho = varImg.shape[1]

    # Creamos el lienzo
    monocromo = np.zeros((alto, ancho, 3))

    # Recorremos imagen pixel a pixel
    for i in range(0, alto):
        for j in range (0, ancho):
            pixel = varImg[i, j]

            red = pixel[0]
            green = pixel[1]
            blue = pixel[2]

            #Pasandole solo el canal R, el órden es BGR
            if (color == 1):
                monocromo[i, j] = [blue, 0, 0]
            elif (color == 2):
                monocromo[i, j] = [0, green, 0]
            elif (color == 3):
                monocromo[i, j] = [0, 0, red]
            elif (color == 10):
                monocromo[i, j] = [0, green, red]
            elif (color == 20):
                monocromo[i, j] = [blue, green, 0]
            elif (color == 30):
                monocromo[i, j] = [blue, 0, red]
            else:
                monocromo[i, j] = [blue, green, red]

    # Realizando la compresión del tensor que hemos generado para convertirlo a imagen
    cv2.imwrite("Resultado.png", monocromo)

    # Leyendo ahora la nueva imagen
    img_R = cv2.imread("Resultado.png")
    
    # Convirtiendo la imagen al formato RGB
    img_R = cv2.cvtColor(img_R, cv2.COLOR_BGR2RGB)
    plt.imshow(img_R)
    plt.show()
```

```python
# Llamado a la función
strImg = 'Img1.jpg'
img = cv2.imread(strImg, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
activarCanales(img, 2)
```

### Problema 2

```python
# Definimos función
def construirImagen(varImgB, varImgG, varImgR):

    # Obteniendo dimensiones
    alto = varImgB.shape[0]
    ancho = varImgB.shape[1]

    # # Creamos el lienzo
    monocromo = np.zeros((alto, ancho, 3))

    # Recorremos imagen pixel a pixel
    for i in range(0, alto):
        for j in range (0, ancho):
            # pixel = varImg[i, j]

            red = varImgR[i, j]
            green = varImgG[i, j]
            blue = varImgB[i, j]

            monocromo[i, j] = [blue, green, red]

    # Realizando la compresión del tensor que hemos generado para convertirlo a imagen
    cv2.imwrite("Resultado.png", monocromo)

    # Leyendo ahora la nueva imagen
    img_R = cv2.imread("Resultado.png")
    
    # Convirtiendo la imagen al formato RGB
    img_R = cv2.cvtColor(img_R, cv2.COLOR_BGR2RGB)
    plt.imshow(img_R)
    plt.show()
```

### Problema 3:
Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir una las imágenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2.

```python
# Definimos función
def deconstruirImagen(varImg):

    # Obteniendo dimensiones
    alto = varImg.shape[0]
    ancho = varImg.shape[1]

    # Creamos los lienzos
    monocromo_red = np.zeros((alto, ancho, 1))
    monocromo_blue = np.zeros((alto, ancho, 1))
    monocromo_green = np.zeros((alto, ancho, 1))

    # Recorremos imagen pixel a pixel
    for i in range(0, alto):
        for j in range (0, ancho):
            pixel = varImg[i, j]

            red = pixel[0]
            green = pixel[1]
            blue = pixel[2]

            monocromo_red[i, j] = [red]
            monocromo_blue[i, j] = [blue]
            monocromo_green[i, j] = [green]

    # Realizando la compresión del tensor que hemos generado para convertirlo a imagen
    cv2.imwrite("Resultado_gray_red.png", monocromo_red)
    cv2.imwrite("Resultado_gray_blue.png", monocromo_blue)
    cv2.imwrite("Resultado_gray_green.png", monocromo_green)
```

```python
# Llamando la función
strImg = 'Img1.jpg'
img = cv2.imread(strImg, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
deconstruirImagen(img)
```

```python
# Verificando resultados con la función del problema 2
strImgB = 'Resultado_gray_blue.png'
strImgR = 'Resultado_gray_red.png'
strImgG = 'Resultado_gray_green.png'
imgB = cv2.imread(strImgB, cv2.IMREAD_GRAYSCALE)
imgG = cv2.imread(strImgG, cv2.IMREAD_GRAYSCALE)
imgR = cv2.imread(strImgR, cv2.IMREAD_GRAYSCALE)
construirImagen(imgB, imgG, imgR)
```

### Problema 4:
Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.


```python
# Definimos función
def generarHistogramas(image_path):
    try:
        # Leer la imagen en escala de grises
        image = plt.imread(image_path)
        grayscale_image = np.mean(image, axis=2)

        # Obtener los canales RGB de la imagen
        red_channel = image[:, :, 0]
        green_channel = image[:, :, 1]
        blue_channel = image[:, :, 2]

        # Calcular los histogramas
        grayscale_histogram, _ = np.histogram(grayscale_image.ravel(), bins=256, range=(0, 256))
        red_histogram, _ = np.histogram(red_channel.ravel(), bins=256, range=(0, 256))
        green_histogram, _ = np.histogram(green_channel.ravel(), bins=256, range=(0, 256))
        blue_histogram, _ = np.histogram(blue_channel.ravel(), bins=256, range=(0, 256))

        # Calcular el valor de la media para cada canal
        grayscale_mean = np.mean(grayscale_image)
        red_mean = np.mean(red_channel)
        green_mean = np.mean(green_channel)
        blue_mean = np.mean(blue_channel)

        # Mostrar los histogramas con la línea vertical que indica la media
        plt.figure(figsize=(12, 8))
        plt.subplot(2, 2, 1)
        plt.plot(grayscale_histogram, color='black')
        plt.axvline(x=grayscale_mean, color='red', linestyle='--', label=f"Media: {grayscale_mean:.2f}")
        plt.title('Histograma Escala de Grises')
        plt.legend()

        plt.subplot(2, 2, 2)
        plt.plot(red_histogram, color='red')
        plt.axvline(x=red_mean, color='black', linestyle='--', label=f"Media: {red_mean:.2f}")
        plt.title('Histograma Canal Rojo')
        plt.legend()

        plt.subplot(2, 2, 3)
        plt.plot(green_histogram, color='green')
        plt.axvline(x=green_mean, color='black', linestyle='--', label=f"Media: {green_mean:.2f}")
        plt.title('Histograma Canal Verde')
        plt.legend()

        plt.subplot(2, 2, 4)
        plt.plot(blue_histogram, color='blue')
        plt.axvline(x=blue_mean, color='black', linestyle='--', label=f"Media: {blue_mean:.2f}")
        plt.title('Histograma Canal Azul')
        plt.legend()

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print("Error al mostrar los histogramas:", e)
```

```python
# Llamando la función
strImg = 'Img1.jpg'
generarHistogramas(strImg)
```
